```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../data/training_sentiment.csv", header=None)[:100000]
df.columns = ["polarity", "id", "date", "query", "author", "body"]
```

```python
df[:5]
```

```python
URL_PATTERN = "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
```

```python
users_hashtags = dict()  # {user: {hashtag: # user used that hashtag}}
users_urls = dict()  # {user: {url: # user used that url}}
users_mentions = dict()  # {user: {mention: # user used that mention}}
urls = dict()
for _, (i, data) in enumerate(df.iterrows()):
    user = data["author"]
    body = data["body"]
    
    hashtags = re.findall("(#\w+)\s?", body)
    urls = re.findall(URL_PATTERN, body)
    mentions = re.findall("@\w+", body)
    
    for h in hashtags:
        users_hashtags.setdefault(user, dict())
        users_hashtags[user].setdefault(h, 0)
        users_hashtags[user][h] += 1
    
    for u in urls:
        users_urls.setdefault(user, dict())
        users_urls[user].setdefault(u, 0)
        users_urls[user][u] += 1
        
    for m in mentions:
        users_mentions.setdefault(user, dict())
        users_mentions[user].setdefault(m, 0)
        users_mentions[user][m] += 1
```

```python
urls = dict()
for u, urs in users_urls.items():
    for ur in urs:
        urls.setdefault(ur, 0)
        urls[ur] += 1
```

```python
{k:v for k,v in urls.items() if v > 1}
```

```python
hashs = dict()
for u, hs in users_hashtags.items():
    for h in hs:
        hashs.setdefault(h, 0)
        hashs[h] += 1
```
