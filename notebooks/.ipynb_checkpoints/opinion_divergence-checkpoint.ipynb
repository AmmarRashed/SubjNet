```python
with open('../snaps/reddit.json', 'r') as f:
    data = json.load(f)
    nxg = json_graph.node_link_graph(data, directed=True)
# bgrs = [nxg.edges[e]["bgr"] for e in nxg.edges]
```

```python
nxg = augment_nodes(nxg, {}, community_detection=True)
```

```python
results, opinion_networks = biased_assimilation_analysis(nxg,
                                                         p=0.005,
                                                         propagators=["random 0.5", "w", "jac", "bgr"],
                                                         T=10)

with open("../results.pkl", "wb") as f:
    pickle.dump(results, f)
    
with open("../data/opinion_networks.pkl", "wb") as f:
    pickle.dump(opinion_networks, f)
```

```python
ntw = opinion_networks["bgr"][-1]
for (f, t) in ntw.edges:
    ntw.edges[(f, t)]['aggreement'] = int(ntw.nodes[f]["opinion"] == ntw.nodes[t]["opinion"])
```

```python
clust = list()
subj = list()
for ix, v in dict(nx.clustering(ntw, weight='aggreement')).items():
    ntw.nodes[ix]["opinion_clustering"] = v
    clust.append(v)
    subj.append(ntw.nodes[ix]["subjectivity"])
```

```python
plt.scatter(clust, subj)
```

```python
from scipy.stats import pearsonr
```

```python
pearsonr(clust, subj)
```

### Similarity trumps over subjectivity
### If you are similar to your friend, he is more likely to change your mind
### and vice versa (baseline if similarity is uniformly 0.5)
